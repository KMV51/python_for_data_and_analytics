```python
# Merging
```

```python
df1 = df[['id','chol','age']].loc[0:4]
df1
```

```python
df2 = df[['id','gender','weight']].iloc[[0, 2, 4], :]
df2
```

```python
df3 = pd.merge(df1, df2, how='inner', left_on='id', right_on='id')
df3
```

```python
df4 = pd.merge(df1, df2, how='left', left_on='id', right_on='id')
df4
```

```python
df5 = pd.merge(df2, df1, how='right', left_on='id', right_on='id')
df5
```

```python
df6 = pd.merge(df2, df1, how='outer', left_on='id', right_on='id')
df6
```
